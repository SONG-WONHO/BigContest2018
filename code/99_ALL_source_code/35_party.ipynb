```python
# module 임포트
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset 로딩
x_load = pd.read_csv('C:\\Users\\zeus_\\Desktop\\champion_data\\test\\test_party.csv')
y_load = pd.read_csv('C:\\Users\\zeus_\\Desktop\\champion_data\\test\\test_activity.csv').sort_values(by='acc_id', ascending=True)
```

```python
x_data = x_load
y_data = y_load.acc_id
```

```python
y_data = y_data.to_frame()
```

```python
y_data.index = y_data.acc_id
```

```python
for i in range(1,9):
    y_data['party_cnt_' + str(i)] = 0
```

```python
y_data = y_data.drop(['acc_id'], axis=1)
```

```python
for i in range(1,9):
    for hashed in x_data[x_data['party_start_week'] == i].hashed:
        for value in hashed.split(','):
            try:
                y_data.loc[value, 'party_cnt_' + str(i)] += 1
            except:
                pass
```

```python
y_data = y_data.groupby('acc_id').mean()
```

```python
y_data.to_csv('C:\\Users\\zeus_\\Desktop\\champion_data\\test\\test_party_post.csv')
```
